## K-means Analysis without tumour stage

```python
import prince
from sklearn import datasets
import pandas as pd
import math

dataframe = dataframe.drop(columns=['stage', 'sub_stage'])
# dataframe = dataframe.drop(columns=['sub_stage'])

# stage_undefined = dataframe['stage']!= 'Undefined'
# dataframe = dataframe[stage_undefined]

def calculate_distance(p1, p2):
    #p1 = [-0.063728, 1.476561]
    #p2 = [-1.008608, -0.821417]
    _distance = math.sqrt( ((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2) )
    return _distance
    
    
X = dataframe
#X = dataframe
X1 = X.copy()
print(X.head())

_mca = prince.MCA(
    n_components=2
)

_mca.fit(X) # same as calling ca.fs_r(1)
#_mca = _mca.transform(X) # same as calling ca.fs_r_sup(df_new) for *another* test set.
print(_mca)
type(_mca)

list_coordinates = _mca.column_coordinates(X)

dataframe.head()
```

```python
ax = _mca.plot_coordinates(
X=X,
ax=None,
figsize=(16, 16),
show_row_points=True,
row_points_size=10,
show_row_labels=False,
show_column_points=True,
column_points_size=30,
show_column_labels=True,
legend_n_cols=1)
```

```python
xx = _mca.transform(X)
```

```python
#Obtenir el valor de la K
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(xx).score(xx) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='random').fit(xx)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Predicting the clusters
labels = kmeans.predict(xx)
# Getting the cluster centers
C = kmeans.cluster_centers_
colores=['blue','red','green','cyan', 'yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(xx.iloc[:, 0], xx.iloc[:, 1], c=asignar,s=60)

ax.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
```

```python
plt.scatter(xx.iloc[:, 0], xx.iloc[:, 1], c=asignar,s=60)

centers = kmeans.cluster_centers_
fig = plt.scatter(centers[:, 0], centers[:, 1], marker='*', c='black', s=200, alpha=0.5)
```

```python
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, xx)
closest
```
